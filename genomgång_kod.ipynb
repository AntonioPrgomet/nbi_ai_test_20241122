# Class Demonstration

* Class: The blueprint for creating objects (in this case, Car).
* Instance: An individual object created from a class.
* Attribute: A characteristic of the instance (like color or make).
* Method: A function defined inside a class that acts on instances (like start_engine).

```python
# Define a class
class Car:
    # Initialize attributes
    def __init__(self, make, color):
        self.make = make        # Instance attribute
        self.color = color      # Instance attribute

    # Define a method
    def start_engine(self):
        print(f"The {self.color} {self.make} engine is now running.")
```

```python
# Create an instance of the class (object)
my_car = Car("Toyota", "red")
```

```python
# Access attributes
print(my_car.make)       # Output: Toyota
print(my_car.color)      # Output: red
```

```python
# Call the method
my_car.start_engine()    # Output: The red Toyota engine is now running.
```

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, money):
        self.balance = self.balance + money
    
    def withdraw(self, money):
        if money > self.balance:
            print('You do not have enough money to take out')
        else:
            self.balance = self.balance - money
```

```python
konto_1 = BankAccount('Johanna', 123)
```

```python
print(konto_1.balance)
konto_1.deposit(7)
print(konto_1.balance)
konto_1.withdraw(90)
print(konto_1.balance)
konto_1.withdraw(100)
```

# A short AI example - Linear Regression

```python
# X = np.random.uniform(low=0.0, high=90, size=1000)
# X
```

```python
# Simluate age
X = np.random.uniform(low=0.0, high=90, size=1000)  # Experience in years

# Simulate salaries with added noise
y = 1000 * X + np.random.normal(15000, 25000, 1000)  # Base salary + noise
y = np.maximum(y, 0)  # Set all negative values to 0 (or use a minimum threshold)

# Scatter plot
fig, ax = plt.subplots()
ax.scatter(X, y, alpha=0.6)
ax.set_title("Simulated Salary Levels vs. Age")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
```

```python
from sklearn.linear_model import LinearRegression
X = X.reshape(-1, 1)

reg = LinearRegression()
reg.fit(X, y)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
X_new = np.array([[20], [40], [70]])
reg.predict(X_new)
```

```python
# Scikit-learn och Tensorflow Ã¤r bland de tvÃ¥ mest anvÃ¤nda AI biblioteken. 
# Scikit-learn anvÃ¤nds vÃ¤ldigt mycket fÃ¶r "standard modeller" inom regression och klassificieringsproblem. 

# Tensorflow anvÃ¤nds fÃ¶r neurala nÃ¤tverk (Deep Learning)
# Tensorflow (Google) och Pytorch (Facebook) Ã¤r konkurrenter. Tensorflow Ã¤r stÃ¶rre (mer anvÃ¤ndare). 
```

#### Plotting data and model

```python
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)  # Create a range of ages for the line
y_pred = reg.predict(X_range)  # Predict salaries for the range of ages

fig, ax = plt.subplots()

ax.scatter(X, y, alpha=0.6, label="Data")
ax.plot(X_range, y_pred, color='red', label="Regression Line")
ax.set_title("Simulated Salary Levels vs. Age")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
```
